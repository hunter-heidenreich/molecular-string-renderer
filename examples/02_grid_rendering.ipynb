# Grid Rendering

This notebook demonstrates how to render multiple molecules in grid layouts.

## Setup and Imports

```python
# Import required modules
from molecular_string_renderer import (
    render_molecules_grid, 
    MolecularRenderer,
    RenderConfig,
)
from IPython.display import Image, display
import time
```

## 1. Basic Grid Rendering

```python
# Simple grid example with common organic molecules
basic_molecules = [
    "CCO",              # Ethanol
    "CC(=O)O",          # Acetic acid  
    "C1=CC=CC=C1",      # Benzene
    "C"                 # Methane
]

basic_legends = [
    "Ethanol",
    "Acetic Acid", 
    "Benzene",
    "Methane"
]

# Create a basic grid - auto-fits to 4x1 layout
print("Creating basic molecular grid...")
grid_image = render_molecules_grid(
    molecular_strings=basic_molecules,
    legends=basic_legends,
    output_path="basic_grid.png"
)

print(f"Grid created! Image size: {grid_image.size}")
display(Image("basic_grid.png"))
```

## 2. Custom Grid Layouts

```python
# Create different grid layouts with more molecules
drug_molecules = [
    "CC(=O)OC1=CC=CC=C1C(=O)O",                    # Aspirin
    "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O",              # Ibuprofen  
    "CN1C=NC2=C1C(=O)N(C(=O)N2C)C",               # Caffeine
    "C1=CC(=C(C=C1CCN)O)O",                       # Dopamine
    "CC(C)(C)NCC(C1=CC(=C(C=C1)O)CO)O",           # Salbutamol
    "CN(C)CCOC1=CC=C(C=C1)C(C)(C)C"               # Diphenhydramine (partial)
]

drug_names = [
    "Aspirin",
    "Ibuprofen", 
    "Caffeine",
    "Dopamine",
    "Salbutamol",
    "Diphenhydramine"
]

print("Creating drug molecule grids with different layouts:")

# 2x3 layout
print("\n2x3 Grid Layout:")
grid_2x3 = render_molecules_grid(
    molecular_strings=drug_molecules,
    legends=drug_names,
    mols_per_row=2,  # Force 2 molecules per row
    mol_size=(250, 200),  # Custom size for each molecule
    output_path="drugs_2x3.png"
)
display(Image("drugs_2x3.png"))

# 3x2 layout  
print("\n3x2 Grid Layout:")
grid_3x2 = render_molecules_grid(
    molecular_strings=drug_molecules,
    legends=drug_names,
    mols_per_row=3,  # Force 3 molecules per row
    mol_size=(200, 150),  # Smaller individual molecules
    output_path="drugs_3x2.png"
)
display(Image("drugs_3x2.png"))
```

## 3. Styled Grids with Custom Configuration

```python
# Create a publication-ready grid with custom styling
amino_acids = [
    "N[C@@H](C)C(=O)O",           # Alanine
    "N[C@@H](CC1=CC=CC=C1)C(=O)O", # Phenylalanine
    "N[C@@H](CC(=O)O)C(=O)O",     # Aspartic acid
    "N[C@@H](CCCNC(=N)N)C(=O)O"   # Arginine
]

amino_acid_names = [
    "Alanine (Ala)",
    "Phenylalanine (Phe)", 
    "Aspartic Acid (Asp)",
    "Arginine (Arg)"
]

# Custom configuration for professional look
publication_config = RenderConfig(
    width=300,
    height=250,
    background_color="#f8f9fa",  # Light gray background
    show_hydrogen=True,             # Explicit hydrogens
    show_carbon=True                # Explicit carbons
)

print("Creating publication-ready amino acid grid...")
amino_grid = render_molecules_grid(
    molecular_strings=amino_acids,
    legends=amino_acid_names,
    mols_per_row=2,
    mol_size=(300, 250),
    render_config=publication_config,
    output_format="png",
    output_path="amino_acids_publication.png"
)

print(f"Publication grid created (size: {amino_grid.size})")
display(Image("amino_acids_publication.png"))
```

```python
# For large datasets, use the object-oriented interface for better performance
renderer = MolecularRenderer(
    render_config=RenderConfig(width=200, height=150, background_color="#f0f0f0")
)

# Create multiple grids efficiently
natural_products = [
    ["CC(C)=CCCC(C)=CCO", "CC(C)=CCC[C@@H](C)C1=CC=C(C)CC1", "CC(C)(C)C1=CC=C(O)C=C1", "C=CCc1ccc(O)c(OC)c1"],  # Terpenes & phenolics
    ["C1=CC=C2C(=C1)C=CC=C2", "C1=CC=C2C(=C1)C3=CC=CC=C3C=C2", "C1=CC=C2C(=C1)C=CC3=CC=CC=C32", "C1=CC2=C(C=C1)C1=CC=CC=C1C=C2"],  # Aromatic hydrocarbons  
    ["O=C1C=CC(=O)C=C1", "O=C1C=CC(=O)C(=C1)O", "COC1=CC(=O)C=CC1=O", "O=C1C=C(O)C(=O)C=C1O"],  # Quinones
    ["CN1C=NC2=C1C(=O)N(C)C(=O)N2C", "CCN1C=NC2=C1C(=O)N(C)C(=O)N2C", "CN1C=NC2=C1C(=O)N(CC)C(=O)N2C", "CN1C=NC2=C1C(=O)NC(=O)N2C"]   # Xanthines
]

categories = ["Terpenes & Phenolics", "Aromatic Hydrocarbons", "Quinones", "Xanthine Derivatives"]

print("Creating natural product library grids...")

for i, (molecules, category) in enumerate(zip(natural_products, categories)):
    print(f"\nProcessing {category}...")
    
    start_time = time.time()
    
    # Use the renderer instance
    grid = renderer.render_grid(
        molecular_strings=molecules,
        legends=[f"Compound {j+1}" for j in range(len(molecules))],
        output_path=f"natural_products_{i+1}.png"
    )
    
    processing_time = time.time() - start_time
    print(f"   {category} grid created in {processing_time:.2f}s")
    print(f"   Grid size: {grid.size}")
    
    # Display the grid
    display(Image(f"natural_products_{i+1}.png"))

print(f"\nAll {len(categories)} natural product grids created successfully!")
```
